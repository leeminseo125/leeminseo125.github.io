- Name Feature에서 Mr. Mrs. Ms. 등으로 기혼여부와 성별을 알아낼 수 있다.
- 이제 이름 데이터는 필요 없기에 열 제거한다.

```python
def extract_title(name):
    if "Mr." in name:
        return 0  # 남성
    elif "Mrs." in name:
        return 1  # 기혼 여성
    elif "Miss." in name:
        return 2  # 미혼 여성
    else:
        return -1  # 기타 (예외 처리용)

# Title 열 생성
df2['Title'] = df2['Name'].apply(extract_title)
df2.drop('Name', axis=1, inplace=True)

df2
```

1. Age Feature의 결측 데이터를 Name Feature에서 알아낸 기혼 여부와 성별을 바탕으로 대체한다.

```python
df2["Age"].fillna(df2.groupby("Title")["Age"].transform("median"), inplace=True)
# Title Feature의 값이 같은 사람들의 중앙값으로 NaN데이터를 각 이름에 해당하는 사람들의 중앙값으로 대체한다.
```

```python
df2.info() # age데이터에 891개의 데이터 생성을 완료하였다.
```

```python
# null 값이 많은 Cabin과 생존자 유추에 필요없는 티켓, 탑승자 아이디는 삭제
df2 = df2.drop(columns=['Cabin'])
df2 = df2.drop(columns=['Ticket'])
df2 = df2.drop(columns=['PassengerId'])


df2
```

```python
# 나이를 알기 위해 정렬 -> 0.42부터 80세로 구성되어있음 
age = df2.Age.unique()
age = np.sort(age)
age
```

```python
# 나이에 따른 분류 함수 정의
def classify_age(age):
    if age <= 20:
        return 0
    elif 20 < age <= 30:
        return 1
    elif 30 < age <= 40:
        return 2
    elif 40 < age <= 50:
        return 3
    else:
        return 4

# Age 열을 기반으로 Age 열 값 자체를 분류 값으로 대체
df2['Age'] = df2['Age'].apply(classify_age)
df2
```

```python
# 학습용 데이터를 위해 test.csv 불러오기
test = pd.read_csv('data/test.csv')
# 생존자를 예측하는 데이터셋이기 때문에 test 데이터에는 Survived column이 없음

# 학습용 데이터로 만들기 위해 copy
train_df = df2.copy()
test_df = test.copy()
```

```python
test_df
```

### 결측지 제거

```python
# test data 데이터 제거
test_df = test_df.drop(columns=['Cabin'])
test_df = test_df.drop(columns=['Name'])
test_df = test_df.drop(columns=['Ticket'])
test_df = test_df.drop(columns=['PassengerId'])
test_df
```

```python
# test data 결측치 확인
test_df.isnull().sum()
```

```python
# Age, Fare 결측치 처리
test_df['Age'] = test_df['Age'].fillna(test_df['Age'].interpolate())
test_df['Fare'] = test_df['Fare'].fillna(test_df['Fare'].median())
test_df
```

```python
test_df.loc[test_df['Sex'] == 'male', 'Sex'] = 0
test_df.loc[test_df['Sex'] == 'female', 'Sex'] = 1
test_df
```

```python
train_df.loc[train_df['Sex'] == 'male', 'Sex'] = 0
train_df.loc[train_df['Sex'] == 'female', 'Sex'] = 1
train_df.loc[train_df['Embarked'] == 'S', 'Embarked'] = 0
train_df.loc[train_df['Embarked'] == 'C', 'Embarked'] = 1
train_df.loc[train_df['Embarked'] == 'Q', 'Embarked'] = 2
train_df
```

```python
test_df.loc[test_df['Embarked'] == 'S', 'Embarked'] = 0
test_df.loc[test_df['Embarked'] == 'C', 'Embarked'] = 1
test_df.loc[test_df['Embarked'] == 'Q', 'Embarked'] = 2
test_df
```

```python
train_df.dtypes
```

```python
# object to int
train_df['Sex'] = train_df['Sex'].astype(int)
test_df['Sex'] = test_df['Sex'].astype(int)
train_df.dtypes
```

```python
train_df['Embarked'] = train_df['Embarked'].astype(int)
test_df['Embarked'] = test_df['Embarked'].astype(int)
test_df.dtypes
```

```python
from sklearn.ensemble import RandomForestClassifier

y = train_df['Survived']

features = ['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare', 'Embarked']
X = pd.get_dummies(train_df[features])
X_test = pd.get_dummies(test_df[features])

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X, y)
predictions = model.predict(X_test)

output = pd.DataFrame({'PassengerId': test.PassengerId, 'Survived': predictions})
output.to_csv('result.csv', index=False)
print("Your result was successfully saved!")
```
